```python
#Clean the array. There is an empty dicctionary

for i in range(0, x.shape[1], 1): #start, stop (stop -1), step
    
    a= x[:,i]
    amp=find_peaks(a)[0]
    amp[amp == np.nan]=0  #True or False   == equal .!= False
    print (amp)

#non numeric feature by 0. Not reduce our dataset, loosing ou information. If we remove our row we have to  memorize the index of the original matrix and compare to it. 
#We prefers no to do this (reduce dataset)

```

```python
vect_period = []
for i in range(0, x.shape[1], 1): #start, stop (stop -1), step
    
    a= x[:,i]
    peaks=find_peaks(a , distance = 10)[0]
    peaks[peaks == np.nan]=0  #True or False .!= False
    periods = np.diff(peaks)*3  #Difference between two pics, so between two values of the arrays  #time is spaced by 3
    period = np.mean(periods)
    vect_period.append(period) #append an element
    print(period)

vect_period = np.array(vect_period) #create an array of vect_period
    
```

```python
vect_amplitud = x.max(axis=0)-x.min(axis=0) #amplitud is the difference between 
```
